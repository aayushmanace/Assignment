# Here we create the the a function to parse the tags to  get real time data

```python
def ParsePrice():
    url = "https://finance.yahoo.com/quote/AAPL/"
    soup = Soup(get(url))
    price = soup.find('div',{'class':'D(ib) Mend(20px)'}).text
    return price
```

```python
for i in range(10):
    print('The Current price of Apple stock is ' + str(ParsePrice()))
```

# Historical Data

```python
url = "https://finance.yahoo.com/quote/AAPL/history?p=AAPL" #Historical Data URL
```

```python
soup = Soup(get(url))
```

```python
soup.find('th',{'class':'Ta(start) W(100px) Fw(400) Py(6px) Pend(6px)'}).text
```

```python
dates = soup.find('td',{'class':'Py(10px) Ta(start) Pend(10px)'})
```

```python
date = [date.find('span').text for date in dates]
date
```

```python
open = soup.find('td',{'class':"Py(10px) Pstart(10px)"})
```

```python
data = [o.find('span').text for o in open]
data
```
